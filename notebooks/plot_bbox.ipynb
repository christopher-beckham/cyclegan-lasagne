```python
# Create figure and axes
fig,ax = plt.subplots(1)

# Display the image
ax.imshow(img)

for i in range(len(js['shapes'])):
    rect = patches.Polygon( np.asarray(js['shapes'][i]['points']), edgecolor='white', facecolor='white')
    ax.add_patch(rect)
plt.show()
```

```python
zeros = np.zeros_like(img, dtype="float32")
```

```python
for i in range(len(js['shapes'])):
    poly = js['shapes'][i]['points']
    poly = np.asarray(poly)
    rr, cc = polygon(poly[:,1], poly[:,0], shape=zeros.shape)
    zeros[rr, cc, :] += 1.
```

```python
plt.imshow(zeros)
```

```python
zeros[ zeros == 0 ] = 0.9
```
